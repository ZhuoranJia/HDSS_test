```python
def update_data(data):
    for sublist in data:
        for i in range(3, len(sublist)):
            if sublist[i]:
                coords = sublist[i].split(',')
                if len(coords) == 3:
                    z_value = coords[-1]
                    if z_value == '4.570':
                        coords[-1] = '9.250'
                    elif z_value == '0.000':
                        coords[-1] = '4.570'
                    elif z_value == '9.250':
                        coords[-1] = '14.000'
                    sublist[i] = ','.join(coords)
    return data
```

```python
new_lines = update_data(lines)
for i in new_lines:
    print(i)
```
