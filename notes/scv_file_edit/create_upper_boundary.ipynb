```python
def update_data(data):
    for sublist in data:
        for i in range(3, len(sublist)):
            if sublist[i]:
                coords = sublist[i].split(',')
                if len(coords) == 3:
                    z_value = coords[-1]
                    if z_value == '4.570':
                        coords[-1] = '9.250'
                    elif z_value == '0.000':
                        coords[-1] = '4.570'
                    elif z_value == '9.250':
                        coords[-1] = '14.000'
                    sublist[i] = ','.join(coords)
    return data
```

```python
new_lines = update_data(lines)
for i in new_lines:
    print(i)
```

```python
with open("C:/Users/SID-DRW/OneDrive - Delft University of Technology/Academic Research/PhD research meeting/year 2/Second_Paper/ifc_file/for_python/cellspace_upper_boundary_withstorey.csv", 'w', newline='',encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerows(new_lines)
```
